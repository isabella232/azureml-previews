```python
from azureml.core import Datastore, Workspace, Dataset
from azureml.data.datapath import DataPath

ws = Workspace.from_config()

print('Workspace name: ' + ws.name, 
      'Azure region: ' + ws.location, 
      'Subscription id: ' + ws.subscription_id, 
      'Resource group: ' + ws.resource_group, sep='\n')
```

**You can create a Delta Lake Dataset using the new method "from_delta_lake()" by passing the Named Datastore, relative-path to "_delta_log" and timestamp**

```python
ds = Datastore.get(ws, "workspaceblobstore")
path_datastore = "delta-table/"
timestamp = "2021-08-26T00:00:00Z"
print('Blobstore:' + ds.account_name,
       'Blobstore Containername: ' + ds.container_name, sep='\n')
delta_lake_by_timestamp_new_for_bs = Dataset.Tabular.from_delta_lake(path=(ds, path_datastore), timestamp_as_of=timestamp)
print(delta_lake_by_timestamp_new_for_bs.to_pandas_dataframe())
```

**You can create a Delta Lake Dataset using the new method "from_delta_lake()" by passing the Named Datastore, relative-path to "_delta_log" and Version**

```python
ds2 = Datastore.get(ws, "workspaceblobstore")
path_datastore = "delta-table/"
version_as_of =0
delta_lake_by_version_new_for_bs = Dataset.Tabular.from_delta_lake(path=(ds2, path_datastore), version_as_of=version_as_of)
print(delta_lake_by_version_new_for_bs.to_pandas_dataframe())
```
